```python
#encoding categorical columns
df.replace({"sex": {"male":0, "female": 1}}, inplace=True)
df.replace({"smoker":{"yes":0, "no":1}}, inplace=True)
df.replace({"region":{"southeast":0, "southwest":1, "northeast":2, "northwest":3
                      }}, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
np.set_printoptions(2)
np.concatenate((y_test.reshape(-1, 1), y_pred.reshape(-1, 1)), axis=1)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(reg.predict(X_train), y_train))
print(mean_absolute_error(reg.predict(X_train), y_train))
```

```python
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```
